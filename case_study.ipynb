```python
# Creating a new dataframe with columns 'Country/Region' and 'Long' to map the corresponding country with their respective latitudes.

df_filtered=pd.read_csv("covid_19_confirmed_v1_lyst1747728690432.csv",usecols=['Country/Region','Long'])
df_filtered
```

```python
dataframe_confirmed['Province/State'] = dataframe_confirmed['Province/State'].fillna('All Provinces')
```

```python
dataframe_confirmed.isnull().any()
```

```python
date_cols=dataframe_confirmed.drop(columns=['Province/State','Country/Region','Lat','Long'])
date_cols
```

```python
date_cols=date_cols.columns.tolist()
date_cols
```

```python
dates=pd.to_datetime(date_cols,format='%m/%d/%y')
dates
```

```python
years=dates.year
years
```

```python
yearly_sums=dataframe_confirmed[date_cols].groupby(years,axis=1).sum()
yearly_sums
```

```python
plt.figure(figsize=(10,7))
x=['2020','2021']
y=[yearly_sums[2020].sum(),yearly_sums[2021].sum()]
bars=plt.bar(x,y)
plt.ylim(0,max(y)*1.1)
plt.ticklabel_format(style='plain',axis='y')
for bar in bars:
    height=bar.get_height()
    plt.text(bar.get_x()+bar.get_width()/2,height,f'{int(height):,}',ha='center',va='bottom',fontsize=12)
plt.title("Yearly total number of Confirmed Cases")
plt.xlabel("Year")
plt.ylabel("Number of cases")
plt.grid(axis='y',linestyle='--',alpha=0.7)
plt.show()
```

The year 2021 has the highest number of cases

```python
months=dates.to_period('M')
monthly_cases=dataframe_confirmed[date_cols].groupby(months,axis=1).sum()
monthly_cases
```

```python
total_monthly_cases=monthly_cases.sum(axis=0)
```

```python
# Monthwise barplot for the year 2020
total_monthly_cases_2020=total_monthly_cases[total_monthly_cases.index.year==2020]
plt.figure(figsize=(28,20))
x=['January','February','March','April','May','June','July','August','September','October','November','December']

bars=plt.bar(x,total_monthly_cases_2020.values,color='orange')
plt.ticklabel_format(style='plain',axis='y')
for bar in bars:
    height=bar.get_height()
    plt.text(bar.get_x()+bar.get_width()/2,height,f'{int(height):,}',ha='center',va='bottom',fontsize=14)
plt.title("Month wise total confirmed cases for the year 2020",fontsize=20)
plt.xlabel("Month",fontsize=18)
plt.ylabel("Total confirmed cases",fontsize=18)
plt.grid(axis='y',linestyle='--',alpha=0.7)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.show()
```

In the year 2020 December month has the most number of confirmed cases and January has the lowest number of confirmed cases. 

```python
# Monthwise barplot for the year 2021
total_monthly_cases_2021=total_monthly_cases[total_monthly_cases.index.year==2021]
plt.figure(figsize=(20,10))
x=['January','February','March','April','May']

bars=plt.bar(x,total_monthly_cases_2021.values,color='purple')
plt.ticklabel_format(style='plain',axis='y')
for bar in bars:
    height=bar.get_height()
    plt.text(bar.get_x()+bar.get_width()/2,height,f'{int(height):,}',ha='center',va='bottom',fontsize=14)
plt.title("Month wise total confirmed cases for the year 2021",fontsize=20)
plt.xlabel("Month",fontsize=18)
plt.ylabel("Total confirmed cases",fontsize=18)
plt.grid(axis='y',linestyle='--',alpha=0.7)
plt.xticks(fontsize=16)
# plt.yticks(fontsize=16)
plt.show()
```

```python
df_deaths_long = dataframe_deaths.melt(
    id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'],   # columns to keep
    var_name='Date',                                              # new column for dates
    value_name='Deaths'                                           # new column for death counts
)

# Convert 'Date' column to datetime format
df_deaths_long['Date'] = pd.to_datetime(df_deaths_long['Date'])

print(df_deaths_long.head())
```

```python
df_recovered_long = dataframe_recovered.melt(
    id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'],
    var_name='Date',
    value_name='Recovered'
)
df_confirmed_long = dataframe_confirmed.melt(
    id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'],
    var_name='Date',
    value_name='Confirmed'
)
```

```python
df_confirmed_long = df_confirmed_long.groupby(['Country/Region', 'Date'])['Confirmed'].sum().reset_index()
df_recovered_long = df_recovered_long.groupby(['Country/Region', 'Date'])['Recovered'].sum().reset_index()
```

```python
# Convert 'Date' in all dataframes to datetime:
df_confirmed_long['Date'] = pd.to_datetime(df_confirmed_long['Date'])
df_deaths_long['Date'] = pd.to_datetime(df_deaths_long['Date'])
df_recovered_long['Date'] = pd.to_datetime(df_recovered_long['Date'])
```

```python
# Merge Confirmed and Deaths
df_merged = pd.merge(df_confirmed_long, df_deaths_long,
                     on=['Country/Region', 'Date'], how='outer')

# Merge the result with Recovered
df_merged = pd.merge(df_merged, df_recovered_long,
                     on=['Country/Region', 'Date'], how='outer')

# Ensure Date is datetime
df_merged['Date'] = pd.to_datetime(df_merged['Date'])
```

```python
df_merged['YearMonth'] = df_merged['Date'].dt.to_period('M')

df_monthly = df_merged.groupby(['Country/Region', 'YearMonth'])[['Confirmed', 'Deaths', 'Recovered']].sum().reset_index()
```
